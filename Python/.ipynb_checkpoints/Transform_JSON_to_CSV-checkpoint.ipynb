## Convert Ecobici dictionary into a pandas data frame and export as csv

```python
import json
 
# Opening JSON file
f = open('../extdata/ecobici-modulos.json', encoding="utf8")
 
# returns JSON object as
# a dictionary
data = json.load(f)
```

```python
data
```

```python
import pandas as pd
```

```python
modules = data['stations']
```

```python
modules
```

```python
pdData = pd.DataFrame(modules)
pdData
```

```python
pdData['location']
```

```python
dataLocation = pd.DataFrame(pdData['location'].values.tolist(), index=pdData.index)
dataLocation
```

```python
pdData['location'].values.tolist()
```

```python
pdDataConcat = pd.concat([pdData, dataLocation], axis = 1)
```

```python
pdDataConcat
```

```python
#Export to csv
pdDataConcat.to_csv('../extdata/modules.csv', index=False)
```
